````python
import pandas as pd

# carregar (se ainda não carregou)
url1 = 'https://raw.githubusercontent.com/camilalaranjeira/python-intermediario/main/fifa-wc/matches_1930_2022.csv'
url2 = 'https://raw.githubusercontent.com/camilalaranjeira/python-intermediario/main/fifa-wc/matches_1991_2023.csv'
wcmen_df   = pd.read_csv(url1)
wcwomen_df = pd.read_csv(url2)
wc = pd.concat((wcwomen_df, wcmen_df)).reset_index(drop=True)

# 1) ver colunas atuais (cole a saída aqui se quiser ajuda)
print("Colunas atuais do dataframe:\n")
print(list(wc.columns))
print("\n---\n")

# 2) lista de colunas desejadas (nomes finais esperados pelo exercício)
desejadas = [
 "time_1","time_2","gols_1","gols_2","data","ano","país_sede",
 "comparecimento","resultado","rodada","gols_1_detalhes","gols_2_detalhes",
 "gols_1_contra","gols_2_contra","gols_1_penalti","gols_2_penalti",
 "cartao_vermelho_1","cartao_vermelho_2","cartao_amarelo_1","cartao_amarelo_2","copa"
]

# 3) para cada desejada, testar alternativas comuns (inglês / sem acento / underscores)
alternativas = {
    "time_1": ["time_1","home_team","home_team_name","team1","team_home"],
    "time_2": ["time_2","away_team","away_team_name","team2","team_away"],
    "gols_1": ["gols_1","home_score","home_goals","score1"],
    "gols_2": ["gols_2","away_score","away_goals","score2"],
    "data": ["data","date","match_date"],
    "ano": ["ano","year"],
    "país_sede": ["país_sede","pais_sede","host_country","country","host"],
    "comparecimento": ["comparecimento","attendance"],
    "resultado": ["resultado","result","winner"],
    "rodada": ["rodada","stage","round"],
    "gols_1_detalhes": ["gols_1_detalhes","home_goal_details","home_goals_detail"],
    "gols_2_detalhes": ["gols_2_detalhes","away_goal_details","away_goals_detail"],
    "gols_1_contra": ["gols_1_contra","home_own_goals","home_owngoal"],
    "gols_2_contra": ["gols_2_contra","away_own_goals","away_owngoal"],
    "gols_1_penalti": ["gols_1_penalti","home_penalties","home_penalty_goals"],
    "gols_2_penalti": ["gols_2_penalti","away_penalties","away_penalty_goals"],
    "cartao_vermelho_1": ["cartao_vermelho_1","home_red_cards","home_red"],
    "cartao_vermelho_2": ["cartao_vermelho_2","away_red_cards","away_red"],
    "cartao_amarelo_1": ["cartao_amarelo_1","home_yellow_cards","home_yellow"],
    "cartao_amarelo_2": ["cartao_amarelo_2","away_yellow_cards","away_yellow"],
    "copa": ["copa","tournament","competition","cup"]
}

# 4) montar mapa de renomeação (source_name -> desired_name)
cols_actual = set(wc.columns)
rename_map = {}
missing = []

for desired in desejadas:
    found = None
    for alt in alternativas.get(desired, [desired]):
        if alt in cols_actual:
            found = alt
            break
        # tentar versão sem acentos e minúscula
        alt_norm = alt.replace("á","a").replace("é","e").replace("í","i").replace("ó","o").replace("ú","u").replace("ã","a").replace("õ","o")
        # checar alt_norm também
        if alt_norm in cols_actual:
            found = alt_norm
            break
    if found:
        # se nome já é o desejado, não precisa renomear
        if found != desired:
            rename_map[found] = desired
    else:
        # coluna não encontrada em nenhuma alternativa -> será criada vazia
        missing.append(desired)

# mostrar o que foi encontrado / mapeado
print("Mapeamento que será aplicado (coluna atual -> coluna desejada):")
for k,v in rename_map.items():
    print(f"  {k} -> {v}")
if missing:
    print("\nColunas não encontradas — serão criadas vazias (NaN):")
    print(" ", missing)
print("\n---\n")

# 5) aplicar renomeação
if rename_map:
    wc = wc.rename(columns=rename_map)

# 6) criar colunas faltantes com NaN
for col in missing:
    wc[col] = pd.NA

# 7) agora fazer as conversões **somente** nas colunas que existem no wc
# STRINGS
colunas_string = [
    "time_1", "time_2", "país_sede", "resultado", "gols_1_detalhes",
    "gols_2_detalhes", "gols_1_contra", "gols_2_contra", "gols_1_penalti",
    "gols_2_penalti", "cartao_vermelho_1", "cartao_vermelho_2",
    "cartao_amarelo_1", "cartao_amarelo_2", "copa"
]
col_str_presentes = [c for c in colunas_string if c in wc.columns]
if col_str_presentes:
    wc[col_str_presentes] = wc[col_str_presentes].astype("string")

# INTEIROS
colunas_int = ["gols_1", "gols_2", "ano", "comparecimento"]
col_int_presentes = [c for c in colunas_int if c in wc.columns]
for c in col_int_presentes:
    # converter com coerção; se tiver NaN, avisamos e preenchemos com 0 antes de int
    wc[c] = pd.to_numeric(wc[c], errors="coerce")
    if wc[c].isna().any():
        print(f"Atenção: coluna '{c}' teve valores não convertíveis (NaN). Preenchendo com 0 para converter para int64.")
        wc[c] = wc[c].fillna(0)
    wc[c] = wc[c].astype("int64")

# DATAS
if "data" in wc.columns:
    wc["data"] = pd.to_datetime(wc["data"], errors="coerce")

# CATEGORIA
if "rodada" in wc.columns:
    wc["rodada"] = wc["rodada"].astype("category")

# 8) salvar e mostrar info
wc.to_csv("wc_formatado.csv", index=False)
print("\nArquivo 'wc_formatado.csv' salvo.\n")
wc.info()
````

#### Q4.
* Crie um dataframe `gols` com duas colunas `país` e `total_gols` com o total de gols marcados por cada país durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).
* Imprima o dataframe `gols` ordenado de forma decrescente, para que os times com mais gols fiquem no topo.

Segue um exemplo ilustrativo com o formato do dataframe resultado:

```
país        total_gols
Brazil      120
Argentina   112
Germany     107
...
```

````python
import pandas as pd

# Gols marcados pelo time da casa (convertendo para inteiro)
home = wc[["home_team", "home_goal"]].copy()
home["home_goal"] = pd.to_numeric(home["home_goal"], errors="coerce").fillna(0).astype(int)
home = home.rename(columns={"home_team": "país", "home_goal": "gols"})

# Gols marcados pelo time visitante (convertendo para inteiro)
away = wc[["away_team", "away_goal"]].copy()
away["away_goal"] = pd.to_numeric(away["away_goal"], errors="coerce").fillna(0).astype(int)
away = away.rename(columns={"away_team": "país", "away_goal": "gols"})

# Junta home + away
gols_all = pd.concat([home, away], ignore_index=True)

# Soma total por país e ordena
gols = (
    gols_all.groupby("país")["gols"]
            .sum()
            .reset_index(name="total_gols")
            .sort_values("total_gols", ascending=False)
)

print(gols)
````

#### Q5.
Qual país tomou mais cartões amarelos somando todas as copas?

Neste exercício você vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observação é uma string represetando uma lista dos cartões amarelos de um único jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16’|1:0|Rosana Gómez', '20’|2:0|Gabriela Chávez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o método genérico `apply` para chamar uma função que você vai criar para transformar uma observação de cartão amarelo em uma contagem de elementos da lista.

Lembre de levar em consideração que muitas observações são `NaN`.

Em seguida faça sua mágica para agrupar as informações por país, acumular os cartões de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
país        cartões amarelos
Argentina   97
England     93
Australia   93
...
```

````python
import pandas as pd
import ast  # Para transformar string de lista em lista real

# --- Função para contar cartões amarelos ---
def contar_cartoes(cartoes):
    if pd.isna(cartoes):
        return 0
    try:
        # Converte string de lista em lista real
        lista = ast.literal_eval(cartoes)
        return len(lista)
    except:
        return 0

# --- Cria colunas de contagem de cartões ---
wc["num_cartoes_1"] = wc["home_yellow_card_long"].apply(contar_cartoes)
wc["num_cartoes_2"] = wc["away_yellow_card_long"].apply(contar_cartoes)

# --- Cartões por país (casa e visitante) ---
home = wc[["home_team", "num_cartoes_1"]].rename(columns={
    "home_team": "país",
    "num_cartoes_1": "cartoes"
})

away = wc[["away_team", "num_cartoes_2"]].rename(columns={
    "away_team": "país",
    "num_cartoes_2": "cartoes"
})

# Junta mandante + visitante
cartoes_all = pd.concat([home, away], ignore_index=True)

# Soma total por país
cartoes_total = (
    cartoes_all.groupby("país")["cartoes"]
               .sum()
               .reset_index()
               .rename(columns={"cartoes": "cartões amarelos"})
               .sort_values("cartões amarelos", ascending=False)
)

print(cartoes_total)
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de pênalti.

Para conseguir essa informação, você precisará trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas também são colunas textuais, onde cada observações apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol está na forma `'Jogador(a) · minuto’'`. Por exemplo:
```
'Alex Morgan · 12’|Rose Lavelle · 20’'
```

Lembre de levar em consideração que muitas observações são `NaN`.

Recomendo criar um dicionário à parte, onde cada chave será um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente será a contagem de ocorrências desses nomes.

Em seguida basta converter o seu dicionário em um dataframe e imprimí-lo na forma (valores não são inventados 😁):
```
jogador(a)      num_gols
Marta           17
Miroslav Klose  16
...
```

````python
import pandas as pd
from collections import defaultdict
import ast

# Lista de colunas de gols
colunas_gols = [
    "home_goal_long",
    "away_goal_long",
    "home_penalty_goal",
    "away_penalty_goal"
]

# Dicionário para contar gols por jogador
gols_jogadores = defaultdict(int)

# Função para processar cada célula
def processar_gols(celula):
    if pd.isna(celula):
        return []
    try:
        # Converte string de lista para lista real, se estiver no formato de string
        lista = ast.literal_eval(celula)
    except:
        lista = [celula]  # caso seja apenas uma string simples
    jogadores = []
    for g in lista:
        # Pegamos só o nome antes do ' · ' se existir
        if '·' in g:
            nome = g.split('·')[0].strip()
        else:
            nome = g.strip()
        jogadores.append(nome)
    return jogadores

# Itera sobre todas as colunas
for col in colunas_gols:
    for celula in wc[col]:
        for jogador in processar_gols(celula):
            gols_jogadores[jogador] += 1

# Converte em dataframe
df_gols_jogadores = pd.DataFrame(
    list(gols_jogadores.items()),
    columns=["jogador(a)", "num_gols"]
)

# Ordena do maior para o menor
df_gols_jogadores = df_gols_jogadores.sort_values("num_gols", ascending=False)

# Mostra top 10
print(df_gols_jogadores.head(10))
````
